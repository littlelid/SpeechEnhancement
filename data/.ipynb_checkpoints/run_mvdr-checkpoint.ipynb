```python
import sys
sys.path.append("..")
import os
import numpy as np
import matplotlib.pyplot as plt
import warnings
from scipy.io import wavfile
import IPython
import pyroomacoustics as pra
from utils import hann, normalize, highpass, linear_2D_array
# Spectrogram figure properties
figsize=(15, 7)        # figure size
fft_size = 512         # fft size for analysis
fft_hop  = 8           # hop between analysis frame
fft_zp = 512           # zero padding
analysis_window = np.concatenate((hann(fft_size), np.zeros(fft_zp)))
t_cut = 0.83           # length in [s] to remove at end of signal (no sound)

# Some simulation parameters
Fs = 8000
t0 = 1./(Fs*np.pi*1e-2)  # starting time function of sinc decay in RIR response
absorption = 0.1
max_order_sim = 2
sigma2_n = 5e-7

# Microphone array design parameters
mic1 = np.array([2, 1.5])   # position
M = 8                       # number of microphones
d = 0.08                    # distance between microphones
phi = 0.                    # angle from horizontal
max_order_design = 1        # maximum image generation used in design
         # array shape
Lg_t = 0.100                # Filter size in seconds
Lg = np.ceil(Lg_t*Fs)       # Filter size in samples
delay = 0.050               # Beamformer delay in seconds

# Define the FFT length
N = 1024


R = linear_2D_array(mic1, M, phi, d) 


path = ''


rate1, signal1 = wavfile.read(path + 'singing_'+str(Fs)+'.wav')
signal1 = np.array(signal1, dtype=float)
signal1 = pra.normalize(signal1)
#signal1 = pra.highpass(signal1, Fs) # 滤掉低于50HZ的语音信号
delay1 = 0.

# The second signal (interferer) is some german speech
rate2, signal2 = wavfile.read(path + 'man1_'+str(Fs)+'.wav')
signal2 = signal2[:len(signal1)] ###
signal2 = np.array(signal2, dtype=float)
signal2 = normalize(signal2)
#signal2 = highpass(signal2, Fs)  # 滤掉低于50HZ的语音信号
delay2 = 1.

```

```python
def spectroplot(Z, N, hop, fs, filename, fdiv=None, vmin= None, vmax=None, cmap=None, interpolation='none', ):
    plt.imshow(
        20 * np.log10(np.abs(Z[:N // 2 + 1, :])),
        aspect='auto',
        origin='lower',
        vmin=vmin, vmax=vmax, cmap=cmap, interpolation=interpolation)

    # label y axis correctly
    plt.ylabel('Freq / Hz')
    yticks = plt.getp(plt.gca(), 'yticks')
    plt.setp(plt.gca(), 'yticklabels', np.array(np.round(yticks / float(N) * fs), dtype=np.int))
    print( np.array(np.round(yticks / float(N) * fs), dtype=np.int) ) 
    # label x axis correctly
    
    plt.xlabel('Time / s')
    xticks = plt.getp(plt.gca(), 'xticks')
    plt.setp(plt.gca(), 'xticklabels', xticks / float(fs) * hop)
    plt.savefig(filename + ".pdf")
    
def plot(F, filename):
    spectroplot(F.T, fft_size+fft_zp, fft_hop, Fs, filename = filename , 
             interpolation='none', cmap=plt.get_cmap('winter'))
```

```python
room_dim = [4, 6]   # 4m X 6m 的房间
room1 = pra.ShoeBox(
    room_dim,
    absorption=absorption,
    fs=Fs,
    t0=t0,
    max_order=max_order_sim,
    sigma2_awgn=sigma2_n)

# Add sources to room
good_source = np.array([1, 4.5])           # 目标信号
normal_interferer = np.array([2.8, 4.3])   # 干扰信号
room1.add_source(good_source, signal=signal1, delay=delay1)
room1.add_source(normal_interferer, signal=signal2, delay=delay2)
```

```python
from beamforming import Beamformer
bf = Beamformer(R, Fs, N=N, Lg=Lg)
room1.add_microphone_array(bf)
room1.compute_rir()
room1.simulate()
```

```python
bf.mvdr_beamformer(room1.sources[0][0:1],
                    room1.sources[1][0:1],
                    sigma2_n*np.eye(int(Lg)*int(M) ), delay=delay)
output_mvdr = bf.filter_out()
output_mvdr = normalize(highpass(output_mvdr, Fs))
```

```python
F = pra.stft(output_mvdr, fft_size, fft_hop, 
          win=analysis_window, 
          zp_back=fft_zp)
plot(F, "MVDR")
```

```python
F = pra.stft(normalize(highpass(bf.signals[bf.M//2], Fs)), fft_size, fft_hop, 
          win=analysis_window, 
          zp_back=fft_zp)
plot(F, 'micro')
```

```python
bf.delay_and_sum_weights(room1.sources[0][0:1],
                    room1.sources[1][0:1],
                    sigma2_n*np.eye(int(Lg)*int(M)) )
bf.filters_from_weights()
output_ds = bf.filter_out()
output_ds = normalize(highpass(output_ds, Fs))
```

```python
F = pra.stft(output_ds, fft_size, fft_hop, 
          win=analysis_window, 
          zp_back=fft_zp)
plot(F, 'DS')
```

```python
F = pra.stft(signal1, fft_size, fft_hop, 
          win=analysis_window, 
          zp_back=fft_zp)
plot(F, 'clean')
```
